# BigQuery Connection Validation

This notebook verifies that your local environment is correctly authenticated to GCP and can query BigQuery public datasets.

```python
from google.cloud import bigquery
import pandas as pd

print("BigQuery client library successfully imported!")
```

## Initialize the Client

If you configured Application Default Credentials using `gcloud auth application-default login`, the client will auto-discover your project and credentials.

```python
# Replace with your actual GCP Project ID if needed, or leave None to auto-detect from default environment credentials
PROJECT_ID = None

try:
    client = bigquery.Client(project=PROJECT_ID)
    print(f"Successfully initialized client for GCP project: '{client.project}'")
except Exception as e:
    print(f"Initialization failed: {e}")
    print("Please check your GCP credentials setup (see README.md).")
```

## Test Query on a Public Dataset

Let's fetch some datasets in the `bigquery-public-data` project to ensure credentials work.

```python
try:
    # List datasets in the public datasets project
    datasets = list(client.list_datasets(project='bigquery-public-data', max_results=5))
    print("Successfully connected to BigQuery public datasets!")
    print("Here are some public datasets available:")
    for dataset in datasets:
        print(f" - {dataset.dataset_id}")
except Exception as e:
    print(f"Failed to run query/list datasets: {e}")
```
