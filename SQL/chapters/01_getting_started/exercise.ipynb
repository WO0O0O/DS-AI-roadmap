# Chapter 1: Getting Started with SQL and BigQuery

### Core Concepts:
* **BigQuery Hierarchy**: Projects contain Datasets, which contain Tables.
* **Client**: Represents the connection to GCP BigQuery.
* **Dataset**: A collection of tables (analogous to a database schema).
* **Table**: Holds rows and columns of data, defined by a schema.
* **Schema**: A list of fields describing columns (`name`, `field_type`, `mode`, `description`).

Let's practice retrieving datasets, inspect a table's schema, and preview row data.

```python
from google.cloud import bigquery

# 1. Initialize the client
client = bigquery.Client()
```

## Task 1: Fetch the Chicago Crime Dataset

We will pull the public dataset `chicago_crime` from the `bigquery-public-data` project.

```python
# Reference the dataset
dataset_ref = client.dataset("chicago_crime", project="bigquery-public-data")

# Fetch the dataset metadata
dataset = client.get_dataset(dataset_ref)

print(f"Successfully fetched dataset '{dataset.dataset_id}' from project '{dataset.project}'")
```

## Task 2: List Tables in the Dataset

Now let's list the tables contained in this dataset.

```python
# List all tables in the dataset
tables = list(client.list_tables(dataset))

print(f"Number of tables in the dataset: {len(tables)}")
for t in tables:
    print(f" - Table Name: {t.table_id}")
```

## Task 3: Inspect Table Schema

Let's fetch the main `crime` table and look at its structure (its schema).

```python
# Reference the 'crime' table
table_ref = dataset_ref.table("crime")

# Fetch the table object
table = client.get_table(table_ref)

# Print out table schema fields
print("Table Schema:")
for field in table.schema:
    print(f"Field: {field.name:<20} | Type: {field.field_type:<10} | Mode: {field.mode:<10}")
```

## Task 4: Preview Data as a Pandas DataFrame

To avoid scanning large amounts of data (which costs money/quota), we can use `list_rows` to fetch a small preview of the table's rows directly without running a SQL query.

```python
# Preview the first 5 rows of the 'crime' table
preview = client.list_rows(table, max_results=5).to_dataframe()

# Display the DataFrame
preview
```
